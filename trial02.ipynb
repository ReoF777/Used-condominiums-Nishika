```python
def convert_wareki_to_seireki(wareki):
    if wareki == wareki:
        if wareki == '戦前':
            wareki = '昭和20年'
        value = wareki[2:-1]
        if value == '元':
            value = 1
        else:
            value = int(value)
    
        if '昭和' in wareki:
            seireki = 1925+value
        elif '平成' in wareki:
            seireki = 1988 + value
        elif '令和' in wareki:
            seireki = 2018+value
    else:
        seireki = wareki
    
    return seireki
```

```python
train['建築年'] = train['建築年'].apply(lambda x: convert_wareki_to_seireki(x))
test['建築年'] = test['建築年'].apply(lambda x: convert_wareki_to_seireki(x))
```

```python
def convert_trade_time(num):
    if num==num:
        year = num[:4]
        month = num[6]
        result = int(year) + 1/4*int(month)
    else:
        result = num
    
    return result
```

```python
train['取引時点'] = train['取引時点'].apply(lambda x: convert_trade_time(x))
test['取引時点'] = test['取引時点'].apply(lambda x: convert_trade_time(x))
```

```python
def normalize_time_station(moyori):
    if moyori == moyori:
        if moyori == '30分?60分':
            moyori = 45
        elif moyori == '1H?1H30':
            moyori = 75
        elif moyori == '2H?':
            moyori = 120
        elif moyori == "1H30?2H":
            moyori = 105
    
    
    return int(moyori)
```

```python
train['最寄駅：距離（分）'] = train['最寄駅：距離（分）'].fillna(0)
test['最寄駅：距離（分）'] = test['最寄駅：距離（分）'].fillna(0)
```

```python
train['最寄駅：距離（分）'] = train['最寄駅：距離（分）'].apply(lambda x: normalize_time_station(x))
test['最寄駅：距離（分）'] = test['最寄駅：距離（分）'].apply(lambda x: normalize_time_station(x))
```
